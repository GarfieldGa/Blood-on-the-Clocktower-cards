## Card Formatting

```python
character = "Shugenja"
url = "https://wiki.bloodontheclocktower.com/"+character
r = requests.get(url)
soup = BeautifulSoup(r.content)

escapes = ''.join([chr(char) for char in range(1, 32) if char != 10])
table = str.maketrans('','',escapes)

text = []
for p_node in soup.body.find_all('p'):
    if (p_node.string and p_node.string[0]=='"') or p_node.get('class','')==['flavour']:
        string = ''
        for child in p_node.children:
            if type(child) == bs4.element.NavigableString:
                string+=child
        string = string.translate(table)
        text.append(string)


front = Image.open('../front_final.png').copy()
im = Image.open('../Townsfolk/jpg/Shugenja.jpg').copy()
X = 314
Y = 214
W = 396
H = 396
im.thumbnail((W,H),Image.Resampling.LANCZOS)
front.paste(im, (X,Y))

# 46, 74, 158
color = (29, 43, 204)
# Call draw Method to add 2D graphics in an image
I1 = ImageDraw.Draw(front)
# Custom font style and font size
nameFont = ImageFont.truetype('../11345-fontps.ttf',48)
X = 358.5
Y = 613.1
name_text = character
# Add Text to an image
_, _, right, bottom = I1.textbbox((0,0),name_text,font=nameFont,align='center')
I1.text(((front.size[0]-right)/2,Y), name_text, fill=color, font=nameFont)
descripFont = ImageFont.truetype('../Cambria Math.ttf',22)
mottoFont = ImageFont.truetype('../Arial-Unicode-Regular.ttf',12,encoding="utf-8")

descrip_text = text[1].strip('"')
motto_text = text[0]

W = 396
H = 164
textbox = Image.new('RGBA', (W-8, H))
I2 = ImageDraw.Draw(textbox)

d_max = 1-3
d_test = 'a'
while (I2.textbbox((0,0),d_test,font=descripFont,align='center')[2]<=W):
    d_test += 'a'
    d_max+=1
m_max = 1-3
m_test = 'a'
while (I2.textbbox((0,0),m_test,font=mottoFont,align='center')[2]<=W):
    m_test += 'a'
    m_max+=1

print(d_max)
print(m_max)
descrip = '\n'.join(map(lambda s: s.strip(), textwrap.wrap(descrip_text,d_max)));
motto = '\n'.join(map(lambda s: s.strip(), textwrap.wrap(motto_text, m_max)));
# descrip = list(map(lambda s: s.strip(), textwrap.wrap(descrip_text,38)));
# motto = list(map(lambda s: s.strip(), textwrap.wrap(motto_text,53)));
# print(repr(motto))

# X,Y=(0,0)
# _,_,_,bottom = I2.textbbox((0,0),descrip[0],font=descripFont,align='center')
# for s in descrip:
#     I2.text((X,Y),s,fill=color,font=descripFont,align='center',anchor='lt')
#     Y += bottom+4
# _,_,_,bottom = I2.textbbox((0,0),motto[0],font=mottoFont,align='center')
# Y+=12
# for s in motto:
#     I2.text((X,Y),s,fill=color,font=mottoFont,align='center',anchor='lt')
#     Y += bottom+4
_,_,right,bottom = I2.multiline_textbbox((0,0),descrip,font=descripFont,align='center')
I2.multiline_text(((W-right)/2,0),descrip,fill=color,font=descripFont,align='center')
_,_,right,_ = I2.multiline_textbbox((0,0),motto,font=mottoFont,align='center')
I2.multiline_text(((W-right)/2,bottom+8),motto,fill=color,font=mottoFont,align='center')
X=314
Y=662
front.paste(textbox, (X+4,Y),mask=textbox)

# Clip image to card size
W = 610
H = 813
left = (1024-W)/2
top = 118
right = (1024+W)/2
bottom = top+813
front = front.crop((left, top, right, bottom))

W = 816
H = 1110
front = front.resize((W,H),Image.Resampling.LANCZOS)
# Display edited image
front.save('../test.png')
```

## Full Code

```python
base_url = "https://wiki.bloodontheclocktower.com/"
categories = ['Demon','Minion','Outsider','Townsfolk','Traveller','Fabled'] 
escapes = ''.join([chr(char) for char in range(1, 32) if char != 10])
table = str.maketrans('','',escapes)
# load fonts
nameFont = ImageFont.truetype('../11345-fontps.ttf',48)
descripFont = ImageFont.truetype('../Cambria Math.ttf',22)
mottoFont = ImageFont.truetype('../cour.ttf',12,encoding="utf-8")
d_max = 1-3
d_test = 'a'
while (I2.textbbox((0,0),d_test,font=descripFont,align='center')[2]<=W-4):
    d_test += 'a'
    d_max+=1
m_max = 1-3
m_test = 'a'
while (I2.textbbox((0,0),m_test,font=mottoFont,align='center')[2]<=W-4):
    m_test += 'a'
    m_max+=1


for i,c in enumerate(categories):
    dir=os.listdir('../'+c+'/jpg/')
    for filename in dir:
        # load card frame
        front = Image.open('../front_final.png').copy()
        im = Image.open('../'+c+'/jpg/'+filename).copy()
        character = filename.split('.')[0]
        save_name = character
        character = character.replace('2','')
        name_text = character
        character = character.replace(' ','_')
        character = character.replace('\'','%27')
        r = requests.get(base_url+character)
        soup = BeautifulSoup(r.content)
        print(character)
        text = []
        for p_node in soup.body.find_all('p'):
            if (p_node.string and p_node.string[0]=='"') or p_node.get('class','')==['flavour']:
                string = ''
                for child in p_node.children:
                    if type(child) == bs4.element.NavigableString:
                        string+=child
                string = string.translate(table)
                text.append(string)

        if len(text)==2:
            descrip_text = text[1].replace('"','')
            motto_text = text[0]
        else:
            descrip_text = text[0].replace('"','')
            motto_text = ''

        X = 314
        Y = 214
        W = 396
        H = 396
        im.thumbnail((W,H),Image.Resampling.LANCZOS)
        front.paste(im, (X,Y))
        # 46, 74, 158
        color = (29, 43, 204)
        if c in ['Demon','Minion']:
            color = (209, 23, 79)
        elif c in ['Fabled']:
            color = (209, 159, 52)
        elif c in ['Traveller']:
            color = (173, 22, 219)
            
        # Call draw Method to add 2D graphics in an image
        I1 = ImageDraw.Draw(front)
        # Custom font style and font size
        X = 358.5
        Y = 613.1
        
        # Add Text to an image
        _, _, right, bottom = I1.textbbox((0,0),name_text,font=nameFont,align='center')
        I1.text(((front.size[0]-right)/2,Y), name_text, fill=color, font=nameFont)
        W = 396
        H = 170
        textbox = Image.new('RGBA', (W-8, H))

        descrip = '\n'.join(map(lambda s: s.strip(), textwrap.wrap(descrip_text,d_max)));
        motto = '\n'.join(map(lambda s: s.strip(), textwrap.wrap(motto_text,m_max)));
        I2 = ImageDraw.Draw(textbox)
        _,_,right,bottom = I2.multiline_textbbox((0,0),descrip,font=descripFont,align='center')
        I2.multiline_text(((W-right)/2,0),descrip,fill=color,font=descripFont,align='center')
        _,_,right,_ = I2.multiline_textbbox((0,0),motto,font=mottoFont,align='center')
        I2.multiline_text(((W-right)/2,bottom+8),motto,fill=color,font=mottoFont,align='center')
        X=314
        Y=662
        front.paste(textbox, (X+4,Y),mask=textbox)
        # Save formatted card front
        front.save('../Fronts/'+save_name+'.png')
```
